```python
import re
import warnings
warnings.filterwarnings("ignore")
```

```python
resumes = {
    "resume_1": "Python, Machine Learning, SQL, Data Analysis",
    "resume_2": "Java, Spring Boot, Microservices",
    "resume_3": "Python, Deep Learning, NLP, SQL",
    "resume_4": "HTML, CSS, JavaScript"
}
```

```python
job_role = "Python Developer"

required_skills = [
    "python",
    "machine learning",
    "sql"
]
```

```python
def clean_text(text):
    text = text.lower()
    text = text.replace(",", "").replace(".", "").replace("!", "")

    return text
```

```python
def match_score(resume_text, skills):
    score = 0
    resume_text = clean_text(resume_text)

    for skill in skills:
        if skill in resume_text:
            score += 1

    return score
```

```python
def decision(score):
    if score >= 2:
        return "Shortlisted"
    else:
        return "Not Suitable"
```

```python
print("AI Resume Screening Agent")
print("Job Role:", job_role)
print("-" * 40)

for name, text in resumes.items():
    score = match_score(text, required_skills)
    result = decision(score)

    print(f"Resume: {name}")
    print(f"Match Score: {score}")
    print(f"Result: {result}")
    print("-" * 40)
```
